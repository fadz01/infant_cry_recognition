```python
# ✅ IMPORT LIBRARIES
import os
import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import librosa.display
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from tqdm import tqdm
import random
import noisereduce as nr

# ✅ SETTINGS
SAMPLE_RATE = 22050
MAX_DURATION = 3
MAX_LENGTH = SAMPLE_RATE * MAX_DURATION
TARGET_CLASSES = ['belly pain', 'burping', 'discomfort', 'hungry', 'tired']
data_dir = r'C:\Users\user\Desktop\NEW DATASET'

os.makedirs("visuals", exist_ok=True)

X = []
y = []

# ✅ CLEANING FUNCTION (Safe Normalize)
# ✅ Clean audio with safe normalization, denoising, silence trimming,noise injection, background removal
def clean_audio(file_path):
    signal, _ = librosa.load(file_path, sr=SAMPLE_RATE)

    max_val = np.max(np.abs(signal))
    if max_val > 0:
        signal = signal / max_val  # Normalize safely
    else:
        print(f"⚠️ Skipped silent file: {file_path}")
        return None

    signal = nr.reduce_noise(y=signal, sr=SAMPLE_RATE)  # Denoising
    signal, _ = librosa.effects.trim(signal)  # Trim silence

    if len(signal) > MAX_LENGTH:
        signal = signal[:MAX_LENGTH]
    else:
        signal = np.pad(signal, (0, MAX_LENGTH - len(signal)))

    return signal


# ✅ AUGMENTATION FUNCTION
def augment_audio(signal):
    aug_list = []

    # Noise Injection
    noise = np.random.normal(0, 0.005, len(signal))
    aug_list.append(signal + noise)  # ✅ Noise injection here

    # Time Stretch
    aug_list.append(librosa.effects.time_stretch(signal, rate=0.9))
    aug_list.append(librosa.effects.time_stretch(signal, rate=1.1))

    # Pitch Shift
    aug_list.append(librosa.effects.pitch_shift(signal, sr=SAMPLE_RATE, n_steps=2))
    aug_list.append(librosa.effects.pitch_shift(signal, sr=SAMPLE_RATE, n_steps=-2))

    # Speed Variation
    def speed_change(y, rate):
        return librosa.resample(y, orig_sr=SAMPLE_RATE, target_sr=int(SAMPLE_RATE * rate))

    aug_list.append(speed_change(signal, 1.2))
    aug_list.append(speed_change(signal, 0.8))

    return aug_list


# ✅ WAVEFORM COMPARISON
# ✅ Save and Show waveform comparison in notebook (Step 3: Visualize Cleaned & Augmented Data)
def save_waveform_comparison(original, cleaned, label, filename):
    fig, axs = plt.subplots(1, 2, figsize=(14, 3))
    fig.suptitle(f"{label} - {filename} (Before vs After)")

    axs[0].set_title("Original")
    librosa.display.waveshow(original, sr=SAMPLE_RATE, ax=axs[0])

    axs[1].set_title("Cleaned")
    librosa.display.waveshow(cleaned, sr=SAMPLE_RATE, ax=axs[1])

    plt.tight_layout()
    plt.show()  # ✅ Display the waveform directly in the notebook
    plt.savefig(f"visuals/waveform_COMPARE_{label}_{filename}.png")
    plt.close()


# ✅ MAIN LOOP
print("🔁 Preprocessing and visualizing...")

for label in TARGET_CLASSES:
    class_path = os.path.join(data_dir, label)
    if not os.path.exists(class_path):
        print(f"❌ Folder not found: {class_path}")
        continue

    for file in tqdm(os.listdir(class_path), desc=label):
        if not file.endswith('.wav'):
            continue

        file_path = os.path.join(class_path, file)
        try:
            original, _ = librosa.load(file_path, sr=SAMPLE_RATE)
            cleaned = clean_audio(file_path)
            if cleaned is None:
                continue

            # Save waveform comparison
            if random.random() < 0.05:
                save_waveform_comparison(original, cleaned, label, file.replace('.wav', ''))
                sf.write(f"visuals/cleaned_{label}_{file}", cleaned, SAMPLE_RATE)

            X.append(cleaned)
            y.append(label)

            # Add augmentations
            for aug in augment_audio(cleaned):
                X.append(aug)
                y.append(label)

        except Exception as e:
            print(f"❌ Error with {file}: {e}")

# ✅ PAD BEFORE SMOTE
print("📊 Padding all signals to fixed length...")
X_fixed = []
for x in X:
    if len(x) > MAX_LENGTH:
        x = x[:MAX_LENGTH]
    else:
        x = np.pad(x, (0, MAX_LENGTH - len(x)))
    X_fixed.append(x)

X_np = np.array(X_fixed)
X_flat = X_np.reshape((X_np.shape[0], -1))
y_encoded = LabelEncoder().fit_transform(y)

# ✅ Save cleaned and augmented data 
X_clean = np.array(X_fixed, dtype=np.float32)
y_clean = np.array(y_encoded)

np.save('X_clean2.npy', X_clean)
np.save('y_clean2.npy', y_clean)


print("✅ Done! Cleaned data saved as X_clean2.npy and y_clean2.npy")

from sklearn.utils.class_weight import compute_class_weight

class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_clean),
    y=y_clean
)

class_weights = dict(enumerate(class_weights_array))
print("✅ Computed class weights:", class_weights)
```

```python
# ✅ 1. Import necessary libraries
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
import os

# ✅ 2. Load cleaned data
SAMPLE_RATE = 22050
X_clean = np.load("X_clean2.npy", mmap_mode='r')
y_clean = np.load("y_clean2.npy")

# ✅ 3. Prepare feature extraction directory
os.makedirs("visuals", exist_ok=True)

# ✅ 4. Define feature extraction function
def extract_features(signal):
    # MFCC + delta + delta-delta
    mfcc = librosa.feature.mfcc(y=signal, sr=SAMPLE_RATE, n_mfcc=13)
    delta = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)
    mfcc_combined = np.vstack([mfcc, delta, delta2])  # (39, T)

    # Mel Spectrogram
    mel = librosa.feature.melspectrogram(y=signal, sr=SAMPLE_RATE)
    mel_db = librosa.power_to_db(mel, ref=np.max)

    # Chroma
    chroma = librosa.feature.chroma_stft(y=signal, sr=SAMPLE_RATE)

    # Spectral Contrast
    contrast = librosa.feature.spectral_contrast(y=signal, sr=SAMPLE_RATE)

    # Crop all to same time frame
    min_T = min(mfcc_combined.shape[1], mel_db.shape[1], chroma.shape[1], contrast.shape[1])
    features = np.vstack([
        mfcc_combined[:, :min_T],
        mel_db[:, :min_T],
        chroma[:, :min_T],
        contrast[:, :min_T]
    ])

    # Normalize
    features = StandardScaler().fit_transform(features)
    return features, mfcc_combined[:, :min_T], mel_db[:, :min_T]

# ✅ 5. Extract features from all signals
X_feat = []
for i in tqdm(range(len(X_clean))):
    features, mfcc_only, mel_only = extract_features(X_clean[i])
    X_feat.append(features)

    # Save plots for the first 5 samples
            # Save plots for the first 5 samples
    if i < 5:
        fig, axs = plt.subplots(1, 2, figsize=(14, 4))
        axs[0].set_title("MFCC + Δ + ΔΔ")
        librosa.display.specshow(mfcc_only, sr=SAMPLE_RATE, x_axis='time', ax=axs[0])
        axs[1].set_title("Mel Spectrogram")
        librosa.display.specshow(mel_only, sr=SAMPLE_RATE, x_axis='time', y_axis='mel', ax=axs[1])
        plt.tight_layout()
        plt.show()
        fig.savefig(f"visuals/features_{i}.png")


# ✅ 6. Pad all to same time dimension
max_len = max(f.shape[1] for f in X_feat)
X_padded = [np.pad(f, ((0, 0), (0, max_len - f.shape[1])), mode='constant') for f in X_feat]
X_final = np.array(X_padded)

# ✅ 7. Save final features and labels
np.save("X_final2.npy", X_final)
np.save("y_final2.npy", y_clean)

print("✅ Feature extraction completed. Saved as X_final.npy and y_final.npy")
```

```python
import numpy as np
from imblearn.over_sampling import RandomOverSampler

# Load original data
X = np.load("X_final2.npy")  # shape: (N, 39, T)
y = np.load("y_final2.npy")

# Flatten X for RandomOverSampler (it needs 2D shape)
X_flat = X.reshape(X.shape[0], -1)

# Apply oversampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_flat, y)

# Reshape X back to original format
X_resampled = X_resampled.reshape(-1, X.shape[1], X.shape[2])

# Save to .npy
np.save("X_balanced.npy", X_resampled)
np.save("y_balanced.npy", y_resampled)

# Check new class distribution
unique, counts = np.unique(y_resampled, return_counts=True)
print("Balanced Class Distribution:")
for cls, count in zip(unique, counts):
    print(f"Class {cls}: {count} samples")
```

```python
import numpy as np
import os

# Load full dataset
X = np.load("X_final2.npy")  # (N, F, T)
y = np.load("y_final2.npy")

# Target per class
TARGET_SAMPLES = 1000
selected_indices = []

# Undersample to 1000 per class
for cls in np.unique(y):
    idx = np.where(y == cls)[0]
    selected = np.random.choice(idx, TARGET_SAMPLES, replace=False)
    selected_indices.extend(selected)

# Shuffle
selected_indices = np.array(selected_indices)
np.random.shuffle(selected_indices)

# Subset the arrays
X_balanced = X[selected_indices]
y_balanced = y[selected_indices]

# Save
np.save("X_balanced2.npy", X_balanced)
np.save("y_balanced2.npy", y_balanced)

# Return final class distribution
unique, counts = np.unique(y_balanced, return_counts=True)
dict(zip(unique, counts))
```
