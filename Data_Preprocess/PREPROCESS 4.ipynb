# STEP 1:Preprocessing (Clean, Normalize, Trim, Pad) 

```python
import numpy as np
np.complex = complex

import librosa
import os
from glob import glob
from tqdm import tqdm

# === CONFIG ===
dataset_path = "Desktop/NEW DATASET 2"
classes = ["belly pain", "burping", "discomfort", "hungry", "tired"]
sample_rate = 22050
target_duration = 1  # seconds
target_len = sample_rate * target_duration

waveforms = []
labels = []

# === PREPROCESS LOOP ===
for idx, cls in enumerate(classes):
    files = glob(os.path.join(dataset_path, cls, "*.wav"))
    print(f"Processing {cls}...")

    for file in tqdm(files):
        y, _ = librosa.load(file, sr=sample_rate)
        y = librosa.util.normalize(y)
        y, _ = librosa.effects.trim(y)

        if len(y) < target_len:
            y = np.pad(y, (0, target_len - len(y)))
        else:
            y = y[:target_len]

        waveforms.append(y)
        labels.append(idx)

# === SAVE ===
np.save("clean_waveforms.npy", np.array(waveforms, dtype=object))
np.save("clean_labels.npy", np.array(labels))
print("✅ Saved clean_waveforms.npy and clean_labels.npy")
```

# STEP 2: Data Augmentation (Stretch, Pitch, Noise)

```python
import numpy as np
import librosa
import random
from tqdm import tqdm

X = np.load("clean_waveforms.npy", allow_pickle=True)
y = np.load("clean_labels.npy")

def augment(y, sr=22050):
    y = y.astype(np.float32)  # make sure it's float
    aug_list = []

    # 1. Pitch shift
    aug_list.append(librosa.effects.pitch_shift(y, sr=sr, n_steps=random.uniform(-2, 2)))

    # 2. Add noise
    noise = np.random.normal(0, 0.005, len(y))
    aug_list.append(y + noise)

    return aug_list

X_aug, y_aug = [], []

for waveform, label in tqdm(zip(X, y), total=len(X)):
    X_aug.append(waveform)
    y_aug.append(label)
    for aug in augment(waveform):
        X_aug.append(aug)
        y_aug.append(label)

np.save("X_augmented.npy", np.array(X_aug, dtype=object))
np.save("y_augmented.npy", np.array(y_aug))

print("✅ Saved X_augmented.npy and y_augmented.npy")
```

# STEP 3:Feature Extraction (MFCC + Mel + Tonnetz)

```python
import numpy as np
np.complex = complex

import librosa
from tqdm import tqdm

X = np.load("X_augmented.npy", allow_pickle=True)
y = np.load("y_augmented.npy")

def extract_features(y, sr=22050, shape=(84, 130)):
    y = np.asarray(y).astype(np.float32)
    y = librosa.util.normalize(y)
    y, _ = librosa.effects.trim(y)

    if len(y) < sr:
        y = np.pad(y, (0, sr - len(y)))
    else:
        y = y[:sr]

    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    delta = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)
    mfcc_stack = np.vstack([mfcc, delta, delta2])  # 39, T

    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=42)
    mel_db = librosa.power_to_db(mel)

    tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(y), sr=sr)
    tonnetz = librosa.util.fix_length(tonnetz, size=mel_db.shape[1], axis=1)
    tonnetz = np.pad(tonnetz, ((0, 42 - tonnetz.shape[0]), (0, 0)), mode='constant')

    mfcc_stack = librosa.util.fix_length(mfcc_stack, size=mel_db.shape[1], axis=1)
    combined = np.vstack([mfcc_stack, tonnetz])  # 84, T

    if combined.shape[1] < shape[1]:
        combined = np.pad(combined, ((0, 0), (0, shape[1] - combined.shape[1])), mode='constant')
    else:
        combined = combined[:, :shape[1]]

    return combined

X_feat = np.array([extract_features(x) for x in tqdm(X)])
np.save("X_final3.npy", X_feat)
np.save("y_final3.npy", y)
print("✅ Saved X_final3.npy and y_final3.npy (ready for training!)")
```

# REPORT IMAGES

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import random

# === Load audio waveforms
X_clean = np.load("clean_waveforms.npy", allow_pickle=True)
y_clean = np.load("clean_labels.npy")
X_aug = np.load("X_augmented.npy", allow_pickle=True)
y_aug = np.load("y_augmented.npy")

class_names = ["belly pain", "burping", "discomfort", "hungry", "tired"]
sample_rate = 22050

# === Plot features in a grid
def plot_features_grid(y_audio, axs_row):
    y_audio = np.asarray(y_audio).astype(np.float32)

    axs_row[0].plot(y_audio)
    axs_row[0].set_title("Waveform")
    axs_row[0].set_xlim(0, len(y_audio))

    mfcc = librosa.feature.mfcc(y=y_audio, sr=sample_rate, n_mfcc=13)
    librosa.display.specshow(mfcc, ax=axs_row[1], x_axis='time')
    axs_row[1].set_title("MFCC")

    mel = librosa.feature.melspectrogram(y=y_audio, sr=sample_rate)
    mel_db = librosa.power_to_db(mel)
    librosa.display.specshow(mel_db, ax=axs_row[2], x_axis='time', y_axis='mel')
    axs_row[2].set_title("Mel Spectrogram")

    tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(y_audio), sr=sample_rate)
    librosa.display.specshow(tonnetz, ax=axs_row[3], x_axis='time')
    axs_row[3].set_title("Tonnetz")

# === Show all classes
for label in range(5):
    class_name = class_names[label]

    orig_idx = random.choice(np.where(y_clean == label)[0])
    aug_idx = random.choice(np.where(y_aug == label)[0])

    y_orig = X_clean[orig_idx]
    y_augm = X_aug[aug_idx]

    fig, axs = plt.subplots(2, 4, figsize=(18, 8))
    plot_features_grid(y_orig, axs[0])
    plot_features_grid(y_augm, axs[1])

    axs[0][0].set_ylabel("Original", fontsize=12)
    axs[1][0].set_ylabel("Augmented", fontsize=12)

    fig.suptitle(f"{class_name.upper()} – Original vs Augmented", fontsize=16)
    plt.tight_layout()
    plt.show()
```
