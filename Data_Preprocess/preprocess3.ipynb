```python
# ✅ IMPORT LIBRARIES
import os
import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import librosa.display
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from tqdm import tqdm
import random
import noisereduce as nr

# ✅ SETTINGS
SAMPLE_RATE = 22050
MAX_DURATION = 3
MAX_LENGTH = SAMPLE_RATE * MAX_DURATION
TARGET_CLASSES = ['belly pain', 'burping', 'discomfort', 'hungry', 'tired']
data_dir = r'C:\Users\user\Desktop\NEW DATASET'

os.makedirs("visuals", exist_ok=True)

X = []
y = []

# ✅ CLEAN AUDIO
def clean_audio(file_path):
    signal, _ = librosa.load(file_path, sr=SAMPLE_RATE)
    if np.max(np.abs(signal)) > 0:
        signal = signal / np.max(np.abs(signal))
    else:
        print(f"⚠️ Skipped silent file: {file_path}")
        return None
    signal = nr.reduce_noise(y=signal, sr=SAMPLE_RATE)
    signal, _ = librosa.effects.trim(signal)
    if len(signal) > MAX_LENGTH:
        signal = signal[:MAX_LENGTH]
    else:
        signal = np.pad(signal, (0, MAX_LENGTH - len(signal)))
    return signal

# ✅ AUGMENTATION
def augment_audio(signal):
    aug_list = []
    noise = np.random.normal(0, 0.005, len(signal))
    aug_list.append(signal + noise)
    aug_list.append(librosa.effects.time_stretch(signal, rate=0.9))
    aug_list.append(librosa.effects.time_stretch(signal, rate=1.1))
    aug_list.append(librosa.effects.pitch_shift(signal, sr=SAMPLE_RATE, n_steps=2))
    aug_list.append(librosa.effects.pitch_shift(signal, sr=SAMPLE_RATE, n_steps=-2))
    aug_list.append(librosa.resample(signal, orig_sr=SAMPLE_RATE, target_sr=int(SAMPLE_RATE * 1.2)))
    aug_list.append(librosa.resample(signal, orig_sr=SAMPLE_RATE, target_sr=int(SAMPLE_RATE * 0.8)))
    return aug_list

# ✅ WAVEFORM VISUALIZATION
def save_waveform_comparison(original, cleaned, label, filename):
    fig, axs = plt.subplots(1, 2, figsize=(14, 3))
    fig.suptitle(f"{label} - {filename} (Before vs After)")
    axs[0].set_title("Original")
    librosa.display.waveshow(original, sr=SAMPLE_RATE, ax=axs[0])
    axs[1].set_title("Cleaned")
    librosa.display.waveshow(cleaned, sr=SAMPLE_RATE, ax=axs[1])
    plt.tight_layout()
    plt.show()
    plt.savefig(f"newvisuals/waveform_COMPARE_{label}_{filename}.png")
    plt.close()

# ✅ LOAD + AUGMENT
print("🔁 Preprocessing and visualizing...")
for label in TARGET_CLASSES:
    class_path = os.path.join(data_dir, label)
    if not os.path.exists(class_path): continue
    for file in tqdm(os.listdir(class_path), desc=label):
        if not file.endswith('.wav'): continue
        path = os.path.join(class_path, file)
        try:
            original, _ = librosa.load(path, sr=SAMPLE_RATE)
            cleaned = clean_audio(path)
            if cleaned is None: continue
            if random.random() < 0.05:
                save_waveform_comparison(original, cleaned, label, file.replace('.wav', ''))
                sf.write(f"newvisuals/cleaned_{label}_{file}", cleaned, SAMPLE_RATE)
            X.append(cleaned)
            y.append(label)
            for aug in augment_audio(cleaned):
                X.append(aug)
                y.append(label)
        except Exception as e:
            print(f"❌ Error with {file}: {e}")

print("📊 Padding all signals...")
X_padded = []
for x in X:
    if len(x) > MAX_LENGTH:
        x = x[:MAX_LENGTH]
    else:
        x = np.pad(x, (0, MAX_LENGTH - len(x)))
    X_padded.append(x)

X_clean = np.array(X_padded, dtype=np.float32)
y_clean = LabelEncoder().fit_transform(y)
np.save("X_clean3.npy", X_clean)
np.save("y_clean3.npy", y_clean)
print("✅ Saved: X_clean3.npy, y_clean3.npy")
```
